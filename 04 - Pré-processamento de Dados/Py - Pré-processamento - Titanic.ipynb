# Titanic

## Conteúdo 

- **Survived**: Indica se o passageiro sobreviveu ao desastre. É atribuído o valor de 0 para aqueles que não sobreviveram, e 1 para quem sobreviveu;
- **Pclass**: Classe na qual o passageiro viajou. É informado 1 para primeira classe; 2 para segunda; e 3 para terceira; Status socioeconômico. 
- **Name**: Nome do passageiro;
- **Sex**: Sexo do passageiro;
- **Age**: Idade do passageiro em anos;
- **SibSp**: Quantidade de irmãos e cônjuges a bordo ;
- **Parch**: Quantidade de pais e filhos a bordo;
- **Ticket**: Número da passagem;
- **Fare**: Preço da passagem;
- **Cabin**: Número da cabine do passageiro;
- **Embarked**: Indica o porto no qual o passageiro embarcou. Há apenas três valores possíveis: Cherbourg, Queenstown e Southampton, indicados pelas letras “C”, “Q” e “S”, respectivamente.

## Carregando as bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import scale, MinMaxScaler
from sklearn.model_selection import train_test_split
```

## Carregando o DataFrame 

```python
df = pd.read_csv("../datasets/titanic/train.csv")
df.head()
```

# Tratamentos

- Remoção/Substituição de valores ausentes ou inválidos (NaN, Null, '')
- Redução da Dimensão
- Seleção/Criação/Sumarização de Atributos
- Discretização/Dimensionamento de Categorias
- Codificação das Categorias

## PassengerId - Remover Coluna

- É um atributo que não interfere na predição
- Apenas identifica unicamente cada linha
- **Remover a coluna**

```python
df = df.drop(columns='PassengerId')
df.head()
```

## Pclass - Conversão de Tipo

- Informam duas coisas
    - A classe social
    - A cabine
- **Tranformar o tipo inteiro para um atributo categórico ordinal**. 

```python
df['Pclass'] = df['Pclass'].astype('category')
df.info()
```

## Age - Substituir Valores Ausentes/Nulos - Mean

- Existem 177 valores NaN dentro do DataFrame
- Substituir os valores NaN pela média das idades
- Observar no nome, o tratamento social que pode indicar possíveis crianças!

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)

print('Valores vazios ', df['Age'].isna().sum())
```

## Names - Transformação/Criação

- **Criar uma nova coluna** contendo o Pronome de Tratamento
- **Remover o Nome**
- Senhor: Sir/ Mr / Don / Jonkheer / Dr / Rev .
- Senhora casada: Madam/ Mrs / Ms / Mme (em frânces) / Countess (condessa).
- Mulher solteira/senhorita/menina: Miss, Mlle (mademoiselle).
- Meninos (formalmente): Master
- Tripulantes : Major / capt / col


```python
def pronomes(x):
    if 'Mr' in x:
        return 'Mr'
    elif 'Don' in x:
        return 'Mr'
    elif 'Mrs' in x:
        return 'Mrs'
    elif 'Ms' in x:
        return 'Mrs'
    elif 'Mme' in x:
        return 'Mrs'
    elif 'Miss' in x:
        return 'Miss'
    elif 'Mlle' in x:
        return 'Miss'
    elif 'Dr' in x:
        return 'Mr'
    elif 'Rev' in x:
        return 'Mr'
    elif 'Master' in x:
        return 'Master'
    elif 'Major' in x:
        return 'Sea'
    elif 'Capt' in x:
        return 'Sea'
    elif 'Jonkheer' in x:
        return 'Mr'
    elif 'Col' in x:
        return 'Sea'
    elif 'Countess' in x:
        return 'Mrs'
    else:
        return 'None'
```

```python
df['Treatment'] = df['Name'].apply(pronomes)
df = df.drop(columns='Name')
df.head()
```

## Sex - Codificação das Categorias - LabelEncoding

- Transforma categoria em valores numéricos

```python
le_sex = LabelEncoder()

df['Sex']  = le_sex.fit_transform(df['Sex'])

df.head()
```

## Embarked - Substituir Valores Ausentes/Nulos - Moda

- Contém 2 valores Nulos
- Como a média não se aplica para valores categóricos, iremos encotrar a categoria mais comum e substituir os valores NaN pela moda. 

```python
 df['Embarked'].value_counts(dropna = False)
```

```python
#Selecionando o calor que mais se repete
moda_embarked = df['Embarked'].value_counts().index[0]
# Selecionando as linhas NaN e a coluna Embarked
df.loc[df['Embarked'].isna(), ['Embarked']] = moda_embarked

print('Total de Valores NaN = ', df['Embarked'].isna().sum())
```

## Embarked - Codificação das Categorias - One-Hot Encoding

- Como os valores não possuem uma ordem, presiamos usar o One-Hot Encoding para criar colunas binárias para cada categoria. 
- **Demos remover uma das colunas para evitar multicolinaridade**
- Depois remover a coluna duplicada

```python
onehot = pd.get_dummies(df['Embarked'], prefix='Embarked', drop_first=True)

df = pd.concat([df, onehot], axis=1)

df = df.drop(columns='Embarked')

df.head()
```

## Treatment - Codificação das Categorias - One-Hot Encoding

- Como os valores não possuem uma ordem, presiamos usar o One-Hot Encoding para criar colunas binárias para cada categoria.
- Demos remover uma das colunas para evitar multicolinaridade
- Depois remover a coluna duplicada

```python
onehot = pd.get_dummies(df['Treatment'], prefix='Treatment', drop_first=True)

df = pd.concat([df, onehot], axis=1)

df = df.drop(columns='Treatment')

df.head()
```

## Cabin - Transformação de Atributos 

- A informação que temos, é a de que passageiros que possuem cabine registradas (1ª classe), possuem mais chances de sobreviver
- Iremos atribuir um valor inteiro de FALSO e VERDADEIRO para caso tenha cabine ou nao. 

```python
df['Cabin'] = df["Cabin"].notnull().astype('int')
df.head()
```

## Ticket - Remoção da Coluna

- O número do ticket não gera nenhum tipo de predição

```python
df = df.drop(columns='Ticket')
df.head()
```

## Age - Discretização dos dados


- Precisamos transformar a Idade em uma variável Categórica

```python
def idades(x):
    if x <=  3:
        return 'Bebê'
    if x <=  10:
        return 'Criança'
    elif x <= 60:
        return 'Adulto'
    else:
        return 'Idoso'

df['Age'] = df['Age'].apply(idades)
df.head()
```

```python
le_age = LabelEncoder()

df['Age']  = le_age.fit_transform(df['Age'])

df.head()
```

## Fare - Dimensionamento de Categorias - Normalização

```python
scaler = MinMaxScaler(feature_range= (0,1)) #Normalização entre 0 e 1
df['Fare'] = scaler.fit_transform(df[['Fare']])
df.head()
```
