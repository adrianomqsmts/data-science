## Visualização

```python
grafo = Graph(edges = [(0,1),(2,3),(0,2),(0,3)], directed = True)
grafo.vs['label'] = ['Fernando', 'Pedro', 'Jose', 'Antonio']
grafo.vs['peso'] = [40,30,30,25]
grafo.es['TipoAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
grafo.es['weight'] = [1,2,1,3]
```

### Definição de Cores para os Vértices

- vertex_color define a sequência de cores dos vércies

```python
grafo.vs['cor'] = ['blue', 'red', 'yellow', 'green']
plot(grafo, bbox=(0,0,300,300), vertex_color = grafo.vs['cor'])
```

### Definição de pesos para as arestas

- edge_width define a largura das arestas de acordo com o peso 

```python
plot(grafo, bbox=(300,300),  edge_width = grafo.es['weight'], vertex_color = grafo.vs['cor'])
```

### Pesos para os vértices

- vertex_size define o tamanho do vértice

```python
plot(
    grafo, bbox=(300,300),
    vertex_size = grafo.vs['peso'],
    edge_width = grafo.es['weight'],
    vertex_color = grafo.vs['cor']
)
```

### Curvatura das arestas

- edge_curved define a curvatura das arestas
    - Pode ser útil para evitar sobreposição de arestas

```python
plot(
    grafo, bbox=(300,300),
    vertex_size = grafo.vs['peso'],
    edge_width = grafo.es['weight'],
    vertex_color = grafo.vs['cor'],
    edge_curved = 0.4
)
```

### Defindo o formato dos vértices

- vertex_shape define o formato do vértice, podendo ser quadrado, um triângulo, etc.

```python
plot(
    grafo, bbox=(300,300),
    vertex_size = grafo.vs['peso'],
    edge_width = grafo.es['weight'],
    vertex_color = grafo.vs['cor'],
    edge_curved = 0.4,
    vertex_shape = 'triangle'    
)
```

## Exibindo os pesos nas arestas

- edge_label define rótulos para as arestas

```python
plot(
    grafo, bbox=(300,300),
    vertex_size = grafo.vs['peso'],
    edge_width = grafo.es['weight'],
    vertex_color = grafo.vs['cor'],
    edge_curved = 0.4,
    vertex_shape = 'triangle',
    edge_label = grafo.es['weight']
)
```

## Caminho

```python
grafo = Graph(edges = [(0,2),(0,1),(1,4),(1,5),(2,3),(6,7),(3,7),(4,7),(5,6)],
                       directed = True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [2,1,2,1,2,1,3,1]
```

```python
plot(grafo, bbox = (0,0,300,300), edge_label = grafo.es['weight'])
```

### Caminho entre os vértices A - H

- get_shortest_paths recebe 
    - Os índices dos vértices que queremos achar o caminho
    - `output = 'vpath'` - retorna os índices dos vértices de todos os caminhos (pode retornar mais de 1)

```python
caminho_vertice = grafo.get_shortest_paths(0,7, output = 'vpath')

for n in caminho_vertice[0]:
    print(grafo.vs[n]['label'])
```

## Menor caminho entre dois vértices

- A função get_shortest_paths retorna as arestas do menos caminho entre dois vértices

```python
caminho_aresta = grafo.get_shortest_paths(0,7, output = 'epath')
caminho_aresta
```

- Mostra o ID dos vértices que fazem parte do caminho

```python
caminho_aresta_id = []
for n in caminho_aresta[0]:
    caminho_aresta_id.append(n)
caminho_aresta_id
```

- Somatório dos pesos (ou distâncias) entre os vértices do caminho

```python
distancia = 0
for e in grafo.es:
    #print(e.index)
    if e.index in caminho_aresta_id:
        distancia += grafo.es[e.index]['weight']
distancia 
```

- Mostra o nome dos vértices que fazem parte do caminho

```python
caminho_nome_vertices = []
for n in caminho_vertice[0]:
    print(grafo.vs[n]['label'])
    caminho_nome_vertices.append(grafo.vs[n]['label'])
caminho_nome_vertices    
```

- Colorir os vértices do caminho

```python
for v in grafo.vs:
    if v['label'] in caminho_nome_vertices:
        v['color'] = 'green'
    else:
        v['color'] = 'gray'
```

- Colorir as arestas do caminho

```python
for e in grafo.es:
    if e.index in caminho_aresta_id:
        e['color'] = 'green'
    else:
        e['color'] = 'gray'
```

## Comunidades

```python
grafo = igraph.load('datasets/Grafo.graphml')
plot(grafo, bbox = (0,0,600,600))
```

### Visualizar comunidades

```python
comunidades = grafo.clusters()
print(comunidades)
```

- Comunidade de cada um dos registros

```python
print(comunidades.membership)
```

- Colorir o gráfico de acordo com a comunidade

```python
# Visualização do grafo
cores = comunidades.membership
# Array de cores para defirmos cores diferentes para cada grupo
cores = np.array(cores)
cores = cores * 20
cores = cores.tolist()
plot(grafo, vertex_color = cores)
```

```python
grafo2 = Graph(edges = [(0,2),(0,1),(1,4),(1,5),(2,3),(6,7),(3,7),(4,7),(5,6)],
                       directed = True)
grafo2.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo2.es['weight'] = [2,1,2,1,2,1,3,1]
# Visualização do grafo
plot(grafo2, bbox = (0,0,300,300))
```

- Comunidades do segundo grafo

```python
comunidades2 = grafo2.clusters()
print(comunidades2)
print(comunidades2.membership)
```

- A função `community_edge_betweenness()` é mais otimizada para visualização das comunidades
- O atributo `optimal_count` exibe o número de clusters
- A função `as_clustering()` Gera as comunidades
- O atributo `membership` retorna as comunidades de cada vértice

```python
c = grafo2.community_edge_betweenness()
print(c)
print('clusters: ', c.optimal_count)
comunidades3 = c.as_clustering()
print('comunidades', comunidades3)
print('comunidades de cada membro: ' ,comunidades3.membership)
```

```python
plot(grafo2, vertex_color = comunidades3.membership)
cores = comunidades3.membership
# Array de cores para defirmos cores diferentes para cada grupo
cores = np.array(cores)
cores = cores * 100
cores = cores.tolist()
plot(grafo2, bbox = (0,0,300,300), vertex_color = cores)
```
